# Exercise 0: Setup the required api keys

Before starting with the other exercises, we will need to provide the API keys (OpenAI, Tavily, Langfuse) to the service. Therefore, execute the following steps:

1. Copy & rename the file `settings.env.template` to `settings.env` in the repositories root folder.
2. Fill in the missing credentials (API keys), that will be provided to you during the workshop.
3. (Optional): Check out the `settings`-object created in `src/core/settings.py`. How and when are the API keys loaded in the environment?
