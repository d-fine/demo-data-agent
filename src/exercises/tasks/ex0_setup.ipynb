# Exercise 0: Setup the required api keys

Before moving on to the other exercises, we need to provide the service with the API keys for OpenAI, Tavily and Langfuse. Therefore, follow these steps:

1. Copy and rename the file `settings.env.template` to `settings.env` in the repository's root folder.
2. Fill in the missing credentials (API keys), which will be provided during the workshop.
3. (Optional): Check out the `settings`-object created in `src/core/settings.py`. Consider how and when the API keys are loaded into the environment.
