# Register Agent Prompt to Langfuse Prompt Management

## Preparation

**Imports**

```python
# ruff: noqa: E402
import sys
from pathlib import Path

import langfuse

# Construct src-path and append it to sys.path
src_path = Path.cwd().parent
sys.path.append(str(src_path))

# Initialize settings-object that loads langfuse keys to the environment
from core.settings import settings  # ty:ignore[unresolved-import]  # noqa: F401

# Initialize the langfuse client
langfuse_client = langfuse.get_client()
```

## Register prompts for all agents

```python
from models.prompts import PromptRole
from prompt_management.chart_summarizer.system_prompt import prompt as chart_summarizer_system_prompt
from prompt_management.chart_summarizer.user_prompt import prompt as chart_summarizer_user_prompt
from prompt_management.executor.system_prompt import prompt as executor_system_prompt
from prompt_management.executor.user_prompt import prompt as executor_user_prompt
from prompt_management.planner.plan_user_prompt import prompt as planner_plan_user_prompt
from prompt_management.planner.replan_user_prompt import prompt as planner_replan_user_prompt
from prompt_management.planner.system_prompt import prompt as planner_system_prompt
from prompt_management.synthesizer.system_prompt import prompt as synthesizer_system_prompt
from prompt_management.synthesizer.user_prompt import prompt as synthesizer_user_prompt
from prompt_management.visualizer.system_prompt import prompt as visualizer_system_prompt
from prompt_management.visualizer.user_prompt import prompt as visualizer_user_prompt
from prompt_management.web_researcher.system_prompt import prompt as web_researcher_system_prompt

# Planner prompt (chat: system, user_plan, user_replan)
langfuse_client.create_prompt(
    name="planner_prompt",
    type="chat",
    prompt=[
        {"role": PromptRole.SYSTEM, "content": planner_system_prompt},
        {"role": PromptRole.USER_PLAN, "content": planner_plan_user_prompt},
        {"role": PromptRole.USER_REPLAN, "content": planner_replan_user_prompt},
    ],
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]

# Executor prompt (chat: system, user_plan)
langfuse_client.create_prompt(
    name="executor_prompt",
    type="chat",
    prompt=[
        {"role": PromptRole.SYSTEM, "content": executor_system_prompt},
        {"role": PromptRole.USER_PLAN, "content": executor_user_prompt},
    ],
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]

# Web researcher prompt (text: system only)
langfuse_client.create_prompt(
    name="web_researcher_prompt",
    type="text",
    prompt=web_researcher_system_prompt,
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]

# Synthesizer prompt (chat: system, user_plan)
langfuse_client.create_prompt(
    name="synthesizer_prompt",
    type="chat",
    prompt=[
        {"role": PromptRole.SYSTEM, "content": synthesizer_system_prompt},
        {"role": PromptRole.USER_PLAN, "content": synthesizer_user_prompt},
    ],
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]

# Visualizer prompt (chat: system, user_plan)
langfuse_client.create_prompt(
    name="visualizer_prompt",
    type="chat",
    prompt=[
        {"role": PromptRole.SYSTEM, "content": visualizer_system_prompt},
        {"role": PromptRole.USER_PLAN, "content": visualizer_user_prompt},
    ],
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]

# Chart summarizer prompt (chat: system, user_plan)
langfuse_client.create_prompt(
    name="chart_summarizer_prompt",
    type="chat",
    prompt=[
        {"role": PromptRole.SYSTEM, "content": chart_summarizer_system_prompt},
        {"role": PromptRole.USER_PLAN, "content": chart_summarizer_user_prompt},
    ],
    labels=["dev_latest"],
)  # ty:ignore[no-matching-overload]
```
