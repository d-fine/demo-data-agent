# Register Agent Prompt to Langfuse Prompt Management

## Preparation

**Imports**

```python
# ruff: noqa: E402
import sys
from pathlib import Path

import langfuse

# Construct src-path and append it to sys.path
src_path = Path.cwd().parent
sys.path.append(str(src_path))

# Initialize settings-object that loads langfuse keys to the environment
from core.settings import settings  # ty:ignore[unresolved-import]  # noqa: F401
```

**Initialize the langfuse client**

```python
langfuse_client = langfuse.get_client()
```

## Create prompt

```python
from models.prompts import PromptRole
from prompt_management.planner.plan_user_prompt import prompt as plan_user_prompt
from prompt_management.planner.replan_user_prompt import prompt as replan_user_prompt
from prompt_management.planner.system_prompt import prompt as planner_prompt

name = "planner_prompt"
prompt_type = "chat"  # Literal["chat"] | None
labels = ["dev_latest"]

prompt = [
        {"role": PromptRole.SYSTEM, "content": planner_prompt},
        {"role": PromptRole.USER_PLAN, "content": plan_user_prompt},
        {"role": PromptRole.USER_REPLAN, "content": replan_user_prompt},
    ]
```

## Register prompt

```python
langfuse_client.create_prompt(
    name=name,
    type=prompt_type,
    prompt=prompt,
    labels=labels,
)  # ty:ignore[no-matching-overload]
```
